```python
def clean_dataset(df):
    assert isinstance(df, pd.DataFrame), "df needs to be a pd.DataFrame"
    df.dropna(inplace=True)
    indices_to_keep = ~df.isin([np.nan, np.inf, -np.inf]).any(1)
    return df[indices_to_keep].astype(np.float64)
```

# Feature selection

###### Feature selection is the method where we select  those features which contribute most to the our predict(dependent) variable. Having irrelavant features in our data can decrease the accuracy of the models and make our model learn based on irrelavant features.

# Feature selection using Kbest method

```python
from sklearn.feature_selection import SelectKBest
sel_five_cols = SelectKBest(mutual_info_classif, k=30)
sel_five_cols.fit(x, y)
X=df[x.columns[sel_five_cols.get_support()]]
```

###### Here we have selected 30 independent features based on Kbest method

```python
X
```

# Handling imbalance

```python
y.value_counts()
```

###### from the value count of target variable we can see that the dataset is imablance. If the dataset is imbalance our output will going to be bias towards the majority class, therefor we need to handle the imbalance.

### Handling imbalance using SMOTE

```python
from imblearn.over_sampling import SMOTE
oversampling=SMOTE()
x_over,y_over=oversampling.fit_sample(X,y)
x_over.shape,y_over.shape
```

###### SMOTE first selects a  minority class instance at  random and find its k nearest minority class neighbours. Then it will synthetically create the observations by finding the mean, untill the minority class observation reaches to majoruty class.

## splitting dataset as 80% of the data as Test part and 20% of data as Train part.

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_over,y_over,test_size=0.3,random_state=0)
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
from collections import Counter
print("Before Smote:",Counter(y))
print("After Smote:",Counter(y_over))
```
